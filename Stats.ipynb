# Считывание данных

Из названий фотографий возможно получить статистику по времени, в которое был произведён снимок. Оценить время года, временной промежуток, день когда возникает больше всего проблем с вывозом. 

```python
dates=[]
l=[]
path='trash/0'
with os.scandir(path) as it:
    for entry in it:
#         print(entry.name)
        if entry.name.endswith(".txt"):
            l+=[entry.path]
            dates+=[[path[-1:], entry.name, datetime.strptime(entry.name[:-4],'%d-%m-%Y_%I-%M-%S_%p')]]
#             dates.append(pd.to_datetime(datetime.strptime(entry.name[:-4],'%d-%m-%Y_%I-%M-%S_%p')))
#             dates.append(pd.to_datetime(entry.name[:-4],format='%d-%m-%Y_%I-%M-%S_%p').to_pydatetime,sort=False)
#             print(path[-1:], datetime.strptime(entry.name[:-4],'%d-%m-%Y_%I-%M-%S_%p'))
dates[:5]           
```

```python
dates_df=pd.DataFrame(dates, columns=['folder','filename','timestamp'])
dates_df[:5]
```
